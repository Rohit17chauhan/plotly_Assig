```python
Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.
```

```python
import plotly.graph_objects as go
```

```python
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic
```

```python

fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic.age,y=titanic.fare,mode='markers'))
fig.show()
```

```python
Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
"""distplot is a function in the seaborn library that allows you to plot a univariate distribution of observations. However, if you'd like to use Plotly Express for similar functionality, you can use the histogram function, which allows you to plot a histogram along with a smoothed kernel density estimate."""
```

```python
import plotly.express as px
import seaborn as sns
import pandas as pd


iris = sns.load_dataset('iris')


fig = px.histogram(iris, x='sepal_length', marginal='rug', title='Distribution of Sepal Length in Iris Dataset')
fig.show()
```
